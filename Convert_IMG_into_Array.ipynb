# **Problem: Border the Image**

*Given an image add borders of width 20px.*

You are required to download the image from this given link and add borders of 20px width and black color (pixel value=0) on all sides of the image. Take care of the data type of the pixels of the image, it's specified here 'int8' for evaluation purposes.

*For submission, a CSV file is required with columns 'c1', 'c2', and 'c3' representing all the three channels of the image and without any index.*

The dimensions of the image here are (420, 315, 3) in channels_last format.

**Note: Don't kill any pixels of the image just add more for the border.**

```python
# Importing all the required python libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image

from numpy import asarray

import matplotlib.image as mpimg
```

```python
#Loading the image

img = mpimg.imread('/content/kitty_image.jpg') #Replace "image.jpg" with the path of your image
plt.imshow(img)
plt.axis('on')
plt.show()
```

```python
#Printing the format, size and mode of the image

image = Image.open("/content/kitty_image.jpg")
print(image.format)
print(image.size)
print(image.mode)
print(type(image))
```

```python
#Adding Border of 20px and Black color
from PIL import Image,ImageOps
image = ImageOps.expand(image, border=20, fill='black')
image.save('border_image.jpeg')
```

```python
#Loading the image
border_image = mpimg.imread('/content/border_image.jpeg')
plt.imshow(border_image)
plt.axis('on')
plt.show()
```

```python
#Converting the image into array
numpy_image_border= asarray(border_image,dtype='int8')
```

```python
#Printing the array
print(numpy_image_border)
```

```python
#Checking the shape of the image
numpy_image_border.shape
```

```python
#Converting into the dataframe
df = pd.DataFrame(numpy_image_border.reshape(-1,3), columns=["c1","c2","c3"])
```

```python
#Printing the dataframe
print(df)
```

```python
#Checking the shape of the dataframe
df.shape
```

```python
#Exporting the image to csv file with dataframe having header but no index
df.to_csv('border_kitty_image.csv', sep=',',index=False)
```

# **Convert the Color Image into Grayscale Image**

```python
#importing required modules and libraries
import matplotlib.pyplot as plt
from skimage import io
from skimage import data
from skimage.color import rgb2gray
from skimage import data
#reading the original image
rgb_img  = io.imread('/content/kitty_image.jpg')
#converting the image into grayscale
gray_img  = rgb2gray(rgb_img)
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()
#setting the axes of the image
ax[0].imshow(rgb_img)
ax[0].set_title("Colored Cat image")
ax[1].imshow(gray_img, cmap=plt.cm.gray)
ax[1].set_title("Black & White Cat image")
#display the processed image
fig.tight_layout()
plt.show()
```
